### Matplotlib Pyplot


```python

import numpy as np 
import matplotlib.pyplot as plt


x_point = np.array([1,2,3,4,5,6])
y_point = np.array([2,4,6,8,10,12])

plt.plot(x_point,y_point,'x')
plt.show()
```

```python

# without line plotting

x_point = np.array([1,3])
y_point = np.array([8,10])

plt.plot(x_point,y_point,'x')
plt.show()
```

```python
z_point = np.array([3,8,1,10])

plt.plot(z_point,marker='o',ms=20,mec='b')
plt.show()
```
